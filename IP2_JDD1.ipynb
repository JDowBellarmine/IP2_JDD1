```python
df['temp_f'] = df.apply(lambda x: (9/5)*x['temp_c']+32,axis=1)
print(df)
```

```python
df=df[['date_time','temp_f','temp_c','dew_point','relative_humidity','wind_speed','visibility','pressure','weather']]
```

```python
df.head()
```

```python
df.date_time = pd.to_datetime(df.date_time)
```

```python
df.dtypes
```

```python
df.head()
```

```python
monthly_avg = df['temp_f'].groupby(df.date_time.dt.month).mean()
```

```python
df1=monthly_avg.to_frame()
```

```python
df1=df1.reset_index()
```

```python
df1=df1.rename(columns = {'index':'','date_time':'month','temp_f':'temp_f'})
```

```python
df1
```

```python
df1['month'] = df1['month'].apply(str)
```

```python
df1.at[0,'month']='January'
df1.at[1,'month']='Febuary'
df1.at[2,'month']='March'
df1.at[3,'month']='April'
df1.at[4,'month']='May'
df1.at[5,'month']='June'
df1.at[6,'month']='July'
df1.at[7,'month']='August'
df1.at[8,'month']='September'
df1.at[9,'month']='October'
df1.at[10,'month']='November'
df1.at[11,'month']='December'
```
